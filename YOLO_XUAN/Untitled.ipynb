```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 1. Vẽ khung lưới ảnh (grid)
fig, ax = plt.subplots(figsize=(8, 6))  # Kích thước ảnh
ax.set_xlim(0, 800)  # Kích thước khung ảnh (x)
ax.set_ylim(0, 600)  # Kích thước khung ảnh (y)

# Tạo một màu nền cho hình ảnh
ax.set_facecolor((0.9, 0.9, 0.9))

# 2. Vẽ lưới (grid cells)
num_cells_x = 8  # Số ô theo chiều ngang
num_cells_y = 6  # Số ô theo chiều dọc
cell_width = 800 / num_cells_x
cell_height = 600 / num_cells_y

# Vẽ các đường lưới
for i in range(num_cells_x + 1):
    ax.axvline(x=i * cell_width, color='black', linestyle='--', lw=0.8)
for j in range(num_cells_y + 1):
    ax.axhline(y=j * cell_height, color='black', linestyle='--', lw=0.8)

# 3. Thêm các bounding box vào grid cells
# Giả sử bạn có các bounding box với thông tin về tọa độ và kích thước trong grid
# (x_min, y_min, width, height) là thông số của bounding box
# Thêm mô tả với class và confidence

bounding_boxes = [
    (1, 2, 3, 2, 0.85, 'Person'),  # (x_min, y_min, width, height, confidence, class)
    (4, 1, 1, 3, 0.92, 'Bicycle'),
]

# Vẽ bounding box
for box in bounding_boxes:
    x_min, y_min, width, height, confidence, class_name = box
    # Chuyển tọa độ của bounding box từ grid sang tọa độ pixel
    x = x_min * cell_width
    y = y_min * cell_height
    w = width * cell_width
    h = height * cell_height
    
    # Vẽ bounding box
    bbox = patches.Rectangle((x, y), w, h, linewidth=2, edgecolor='red', facecolor='none')
    ax.add_patch(bbox)
    
    # Thêm thông tin mô tả cho bounding box
    ax.text(x, y - 10, f'{class_name}: {confidence:.2f}', color='red', fontsize=10, ha='left')

# 4. Hiển thị hình ảnh
ax.set_title('YOLOv8')
plt.gca().invert_yaxis()  # Đảo ngược trục y để giống như hệ tọa độ hình ảnh
plt.show()
```
